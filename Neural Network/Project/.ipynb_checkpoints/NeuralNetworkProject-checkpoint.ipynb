**Part 1.2 a - Split the data into X & Y..**

```python
X = data.drop(columns=['Signal_Strength'])
Y = data['Signal_Strength']
```

```python
X.head()
```

```python
Y.head()
```

**Part 1.2 b - Split the data into train & test with
70:30 proportion.**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

**Part 1.2 c - Print shape of all the 4 variables and
verify if train and test data is in sync.**

```python
print("shape of X train {}, Y train {}".format(X_train.shape,Y_train.shape))
```

```python
print("shape of X test {}, Y test {}".format(X_test.shape,Y_test.shape))
```

X train and Y train has same no of records

X test and Y test has same no of records 

train has extactly 70% of records 951 out of overall 1359

test has extactly 30% of records 408 out of overall 1359


**Part 1.2 d - Normalise the train and test data with
appropriate method**

```python
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
X_train_norm
```

```python
X_test_norm
```

Normalised both train and test data using Z-score normalisation

**Part 1.2 e - Transform Labels into format
acceptable by Neural Network**

```python
encoder = OneHotEncoder(sparse=False)
Y_train_encoded = Y_train.values.reshape(-1, 1)
Y_test_encoded = Y_test.values.reshape(-1, 1)

Y_train_encoded = encoder.fit_transform(Y_train_encoded)
Y_test_encoded = encoder.transform(Y_test_encoded)
```

```python
Y_test_encoded
```

**Part 1.3 a - Design a Neural Network to train a classifier.**

```python
model = Sequential([
    Dense(16, activation='relu', input_shape=(11,)),
    Dense(16, activation='relu'),
    Dense(6, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


model.summary()
```

**Part 1.3 b - Train the classifier using previously designed
Architecture**

```python
history = model.fit(X_train_norm, Y_train_encoded, 
                    batch_size=32, epochs=100, 
                    validation_data=(X_test_norm, Y_test_encoded))
```

**Part 1.3 c - Plot 2 separate visuals
 i) Training Loss and Validation Loss
 ii) Training Accuracy and Validation Accuracy**

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

**Part 1.3 d - Design new architecture/update existing
architecture in attempt to improve the performance of the
model.**

```python
model2 = Sequential([
    Dense(256, activation='relu', input_shape=(11,)),
    Dropout(0.5),
    Dense(512, activation='relu'),  
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dense(6, activation='softmax')
])


model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


model2.summary()
```

```python
history2 = model2.fit(X_train_norm, Y_train_encoded, 
                    batch_size=32, epochs=100, 
                    validation_data=(X_test_norm, Y_test_encoded))
```

**Part 1.3 e - Plot visuals as in Q3.C and share
insights about difference observed in both the
models.**

```python
plt.plot(history2.history['loss'], label='Training Loss')
plt.plot(history2.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss(udpated design)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history2.history['accuracy'], label='Training Accuracy')
plt.plot(history2.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy(udpated design)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
