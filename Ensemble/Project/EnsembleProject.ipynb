**2 c Create a function that will accept a DataFrame as input
and return pie-charts for all theappropriate Categorical
features. Clearly show percentage distribution in the
pie-chart.**

```python
def distribution_pie_charts(data):

    categorical_columns = data.select_dtypes(include=['object']).columns
    for column in categorical_columns:
        values = data[column].value_counts()
        labels = values.index
        percentages = (values / values.sum()) * 100

        plt.figure(figsize=(4, 4))
        plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=90)
        plt.title(f'Distribution Pie Chart for {column}')
        plt.show()
```

```python
distribution_pie_charts(data)
```
